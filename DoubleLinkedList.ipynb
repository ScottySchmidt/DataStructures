```python
# Initialise the Node
class Node:
    def __init__(self, data):
        self.item = data
        self.next = None
        self.prev = None
        
# Class for doubly Linked List
class doublyLinkedList:
    def __init__(self):
        self.head = None
        
    # Insert Element to Empty list
    def InsertToEmptyList(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("The list is empty")
            
    # Insert element at the end
    def InsertToEnd(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
            return
        n = self.head
        while n.next is not None:
            n = n.next
        new_node = Node(data)
        n.next = new_node
        new_node.prev = n
        
    #INSERT BEFORE:     
    def insert_before_item(self, x, data):
        if self.head is None:
            print("List is empty")
            return
        else:
            n = self.head
            while n is not None:
                if n.item == x:
                    break
                n = n.next
            if n is None:
                print("item not in the list")
            else:
                new_node = Node(data)
                new_node.next = n
                new_node.prev = n.prev
                if n.prev is not None:
                    n.prev.next = new_node
                n.prev = new_node
        
    #INSERT AFTER:     
    def insert_after_item(self, x, data):
        n = self.head
        print(n.next)
        while n is not None:
            if n.item == x:
                break
            n = n.next
        if n is None:
            print("item not in the list")
        else:
            new_node = Node(data)
            new_node.next = n.next
            n.next = new_node
    
    #DELETE ELEMENT BY VALUE:
    def deleteNode(self, position):
        if self.head is None:
            return
        if position == 0:
            self.head = self.head.next
            return self.head
        index = 0
        current = self.head
        prev = self.head
        temp = self.head
        while current is not None:
            if index == position:
                temp = current.next
                break
            prev = current
            current = current.next
            index += 1
        prev.next = temp
        return prev

    # Delete the elements from the start
    def DeleteAtStart(self):
        if self.head is None:
            print("The Linked list is empty, no element to delete")
            return 
        if self.head.next is None:
            self.head = None
            return
        self.head = self.head.next
        self.start_prev = None;
               
    # Display List:
    def Display(self):
        if self.head is None:
            print("The list is empty")
            return
        else:
            n = self.head
            while n is not None:
                print("Element is: ", n.item)
                n = n.next
        print("\n")
```

```python
# Create a new Doubly Linked List
dList1 = doublyLinkedList()

# Insert the element to empty list
dList1.InsertToEnd("Sam")
dList1.InsertToEnd("Harry")
dList1.InsertToEnd("Dick")
dList1.InsertToEnd("Tom")
dList1.Display()

#Insert Before:
dList1.insert_before_item("Harry", "BEFORE_TEST")  #Insert Sue before Sam:
dList1.insert_before_item("Tom", "Bill") #Insert Bill before Tom:
dList1.insert_before_item("Sam", "Sue")  #Insert Sue before Sam:

#Insert After:
dList1.insert_after_item("Tom", "AFTER_TEST") #After Tom Test Only:

dList1.Display()

#DELETE Nodes: 
print("Deleting..")
dList1.deleteNode(0) 
dList1.deleteNode(3) 

dList1.Display()
```
