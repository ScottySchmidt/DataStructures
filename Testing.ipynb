### Question 1 
Write a search method search(……) with four parameters: the search array, the target, the start
subscript, and the finish subscript. The last two parameters indicate the part of the array that
should be searched. If the target is found, the method returns the index of the first occurrence of
the target in the array. If the target is not found, the method returns -1. Your method should catch
or throw exceptions where warranted.

```python
def search(arr, target, start, finish):
    for i in range(start, finish):
        if arr[i]==target:
            return i
    else:
        return -1

arr1 = [5, 3, 1, 2] 
test1=search( arr1, 6, 0, len(arr1) )

arr2 = [5, 3, 1, 2, 6, 2] 
test2=search( arr2, 6, 0, len(arr2) )

arr3 = [5, 3, 1, 2, 6, 2] 
test3=search(arr3, 5, 0, len(arr3) )

print(test1) #correctly returns -1 not found
print(test2) #correctly returns index 4
print(test3) #correctly returns index 0
```

(the second parameter) in an array (the first parameter) or −1 if the target is not found.
Modify the test(s) in the listing to verify a method that finds the last occurrence of a target
element in an array.

```python
def arraySearch(arr, target, start, finish):
    num=-1
    for i in range(start, finish):
        if arr[i]==target:
            num=i
    return num 

arr1 = [5, 3, 1, 2, 3, 5] 
test1=arraySearch( arr1, 3, 0, len(arr1) ) #correctly returns 4 index which is the LAST occurance
print(test1)
```
